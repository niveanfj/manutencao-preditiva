##### Analise dos dados:

As colunas 'udi' e 'product_id' possuem itens unicos, logo serão removidos para não interferir nos processos posteriores:

```python
print(F"Quantidade de itens na coluna 'udi': {len(df['udi'].unique())}")
print(F"Quantidade de itens na coluna 'product_id': {len(df['product_id'].unique())}")
print(F"Total de itens no dataset: {len(df)}")
```

```python
df = df.drop(['udi', 'product_id'], axis=1)
df.head()
```
